```python
interval = int(input("Enter a limit value to count Prime Numbers : "))
prime_numbers = []
for i in range(interval, 1, -1) :
    count = 0
    for ii in range(2, i) :
        if not i % ii :
            count += 1
    if not count :
        prime_numbers.append(i) 
print(prime_numbers)      
```
